```python
import pandas
#On charge le fichier excel
df = pandas.read_excel("resultat20221.xlsx") 
df.drop(columns=df.columns[df.columns == ""], inplace=True)

#On supprime les colonnes inutiles 
df.drop(columns=["Libellé du département","Inscrits","Votants","% Vot/Ins","Libellé de la circonscription","Code du b.vote","% Voix/Ins.1","% Voix/Exp.1","% Voix/Ins.2","% Voix/Exp.2","% Voix/Ins.3","% Voix/Exp.3","% Voix/Ins.4","% Voix/Exp.4","% Voix/Ins.5","% Voix/Exp.5","% Voix/Ins.6","% Voix/Exp.6","% Voix/Ins.7","% Voix/Exp.7","% Voix/Ins.8","% Voix/Exp.8","% Voix/Ins.9","% Voix/Exp.9","% Voix/Ins.10","% Voix/Ins.11","% Voix/Exp.11","% Voix/Exp.10","Code de la circonscription","Code de la commune",'% Voix/Ins', 'Sexe','Sexe.1','Sexe.2','Sexe.3','Sexe.4','Sexe.5','Sexe.6','Sexe.7','Sexe.8','Sexe.9', 'Exprimés', "N°Panneau","N°Panneau.1","N°Panneau.2","N°Panneau.3","N°Panneau.4","N°Panneau.5","N°Panneau.6","N°Panneau.7","N°Panneau.8","% Voix/Exp","% Blancs/Ins","% Blancs/Vot","% Nuls/Ins","% Nuls/Vot","% Exp/Ins","% Exp/Vot","% Abs/Ins","d","N°Panneau.9","Sexe.10"], inplace=True)

#On regroupe les colonnes Abstentions,Blancs et Nuls pour faire une seule colonne
df['null'] = df['Abstentions'] + df['Blancs'] + df['Nuls']
df.drop(columns=['Abstentions', 'Blancs', 'Nuls'], inplace=True)


df.head(5)
```

```python
#on récupère la colonne voix qui a le plus de voix dans chaque ligne
colonnes_Voix = df.filter(like='Voix')
max_values = colonnes_Voix.max(axis=1)
idxmax_values = colonnes_Voix.idxmax(axis=1)

#on récupère la valeure de la colonne nom qui est juste avant chaque colonne récupérer ci-dessus pour créer une colonnes gagnant(e) dans chaque ligne et y ajouter cette veleure 
df['gagnant(e)'] = ''
for i, maxValCol in enumerate(idxmax_values):
    liste_colonnes = df.columns
    preceding_column_id = df.columns.get_loc(maxValCol) - 1
    precedingg_column_id = df.columns.get_loc(maxValCol) - 2
    ligne = i
    colonne_a_modifier = liste_colonnes[preceding_column_id]
    colonne_a_modifier2 = liste_colonnes[precedingg_column_id]
    nouvelle_valeur = df.loc[ligne, colonne_a_modifier]
    nouvelle_valeur2 = df.loc[ligne, colonne_a_modifier2]
    #print(nouvelle_valeur)
    #print(maxValCol)
    df.loc[i, 'gagnant(e)'] = str(nouvelle_valeur) + ' ' + str(nouvelle_valeur2)

    
```

```python
#On supprime toutes les colonnes prénom
colonnes_prenom = df.filter(like='Prénom')
df = df.drop(columns=colonnes_prenom)
df.head(5)
```

```python
#On associe un candidat à une orientation politique
orientations_politiques = {
    "ARTHAUD": "gauche",
    "HIDALGO": "gauche",
    "JADOT": "gauche",
    "LASSALLE": "gauche",
    "MÉLENCHON": "gauche",
    "POUTOU": "gauche",
    "ROUSSEL": "gauche",
    "DUPONT-AIGNAN": "droite",
    "LE PEN": "droite",
    "PÉCRESSE": "droite",
    "ZEMMOUR": "droite",
    "MACRON": "milieu",

}

# Filtrer les colonnes contenant 'Nom'
colonnes_nom = df.filter(like='Nom')



# Créer un dictionnaire de correspondance pour le renommage des colonnes
column_mapping = {}
for colonne in colonnes_nom.columns:
    candidat = df.at[1, colonne]
    if candidat in orientations_politiques:
        orientation = orientations_politiques[candidat]
        column_mapping[colonne] = orientation




# Convertir les ensembles en listes
column_mapping = {str(k): str(v) for k, v in column_mapping.items()}

# Renommer les colonnes 'Nom...' par l'orientation politique du nom qui y est stocké (voir ci-dessous)
df.rename(columns=column_mapping, inplace=True)
df.head(5)

```

```python
colonnes = df.columns.tolist()

#Ici on regroupe toutes les colonnes droite, gauche et centre pour les regouper entre elle et aditionner leurs nombres de voix

# Trouver les index de toutes les colonnes ayant le nom "gauche"
index_colonnes_gauche = [i for i, nom_colonne in enumerate(colonnes) if nom_colonne == 'gauche']
df['la gauche'] = 0
for colonne_gauche in index_colonnes_gauche:
    liste_colonnes = df.columns
    nom_colonne_recherchee = liste_colonnes[colonne_gauche + 1]
    df['la gauche'] += df[nom_colonne_recherchee]

df.drop(df.columns[index_colonnes_gauche], axis=1, inplace=True)

    
# Trouver les index de toutes les colonnes ayant le nom "droite"
index_colonnes_droite = [i for i, nom_colonne in enumerate(df.columns) if nom_colonne == 'droite']
df['la droite'] = 0
for colonne_droite in index_colonnes_droite:
    liste_colonnes = df.columns
    nom_colonne_recherchee = liste_colonnes[colonne_droite + 1]
    df['la droite'] += df[nom_colonne_recherchee]
    
df.drop(df.columns[index_colonnes_droite], axis=1, inplace=True)


# Trouver les index de toutes les colonnes ayant le nom "milieu"
index_colonnes_centre = [i for i, nom_colonne in enumerate(colonnes) if nom_colonne == 'milieu']
index_colonne_milieu = df.columns.get_loc('milieu')
nom_colonne_recherchee = liste_colonnes[index_colonne_milieu + 1]
df['milieu'] = df[nom_colonne_recherchee]

# On supprime toutes les colonnes avec le nom voix (elles ne servent plus)
df.drop(columns=["Voix","Voix.1","Voix.2","Voix.3","Voix.4","Voix.5","Voix.6","Voix.7","Voix.8","Voix.9","Voix.10","Voix.11"], inplace=True)

df.head(1)
```
