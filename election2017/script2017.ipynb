```python
import pandas as pd


# Charger les données à partir du fichier Excel
data = pd.read_csv('DATA_election2017_original.csv', sep=';', encoding='utf-8')
data
```

```python
for i in data.columns:
    print(i)
```

```python
orientations_politiques = {
    "DUPONT-AIGNAN": "droite",
    "LE PEN": "droite",
    "MACRON": "milieu",
    "HAMON": "gauche",
    "ARTHAUD": "gauche",
    "POUTOU": "gauche",
    "CHEMINADE": "milieu",
    "LASSALLE": "gauche",
    "ASSELINEAU": "droite",
    "FILLON": "droite",
    "MÉLENCHON": "gauche"
}
condidates_array = [voix.split("_")[1] for voix in data.columns if "Voix" in voix and "Ins" not in voix and "Exp" not in voix]
def getWinner(df):
    maximum = 0
    condidate = False
    
    partie = {
        "droite" : 0,
        "gauche" : 0,
        "milieu" : 0
    }
     
    for i in condidates_array:
        if maximum < df["Voix_"+i]:
            maximum = df["Voix_"+i]
            condidate = df["Nom_"+i]
        
        for p in partie:
            if orientations_politiques[df["Nom_"+i]] == p : 
                partie[p] = partie[p] + df["Voix_"+i]
        
    df["Gagnant"] = condidate
    df["Partie gagnante"] = max(partie, key=partie.get)
    for p in partie:
        df[p] = partie[p]
            
    return df
```

```python
data = data.apply(getWinner, axis=1)[["Code_de_la_commune", "Libellé_de_la_commune", "Gagnant", "Partie gagnante", "droite", "gauche","milieu"]]
```

```python

# Créer un dictionnaire pour les orientations politiques
orientations_politiques = {
    "DUPONT-AIGNAN": "droite",
    "LE PEN": "droite",
    "MACRON": "milieu",
    "HAMON": "gauche",
    "ARTHAUD": "gauche",
    "POUTOU": "gauche",
    "CHEMINADE": "milieu",
    "LASSALLE": "gauche",
    "ASSELINEAU": "droite",
    "FILLON": "droite",
    "MÉLENCHON": "gauche"
}

# Fonction pour déterminer l'orientation politique en fonction du nom du candidat
def determiner_orientation(nom_candidat):
    return orientations_politiques.get(nom_candidat, "inconnu")

# Appliquer la fonction determiner_orientation pour chaque candidat
data['Orientation'] = data['Nom_0'].apply(determiner_orientation)

# Grouper les données par commune et orientation politique, puis calculer les totaux des voix
grouped = data.groupby(['Libellé_de_la_commune', 'Orientation'])['Voix_0'].sum().reset_index()

# Pivotez les données pour avoir une colonne par orientation
pivot_table = grouped.pivot(index='Libellé_de_la_commune', columns='Orientation', values='Voix_0').reset_index()

# Ajouter une colonne pour le candidat gagnant
pivot_table['Candidat_Gagnant'] = pivot_table[['droite', 'gauche', 'milieu']].idxmax(axis=1)

# Afficher les résultats
print(pivot_table)
```
